### Data cleaning 

```python
 # important columns----->
    #genres
    #id
    #keywords
    #title 
    #overview
    #relcast
    #crew
```

```python
movies=movies[["movie_id","title","overview","genres","keywords","cast","crew"]]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0]
```

```python
movies.genres[0]
```

```python
import ast
ast.literal_eval(movies.genres[0])
```

```python
def fun(generes):
    genre=[]
    for i in ast.literal_eval(generes):
        genre.append(i['name'])
    return genre
```

```python
movies["genres"]=movies["genres"].apply(fun)
```

```python
movies["keywords"]=movies["keywords"].apply(fun)
```

```python
def names(n):
    name=[]
    count=0
    for i in ast.literal_eval(n):
        if count!=3:
            name.append(i["name"])
        else:
                break
    return name
```

```python
movies["cast"]=movies["cast"].apply(names)
```

```python
def fetch_director(n):
    director=[]
    count=0
    for i in ast.literal_eval(n):
        if i["job"]=="Director":
            director.append(i["name"])
            break
    return director
```

```python
movies["crew"]=movies["crew"].apply(fetch_director)
```

```python
movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# removing spaces between words
```

```python
movies["genres"]=movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies["keywords"]=movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies["tags"]=movies["overview"]+movies["genres"]+movies["keywords"]+movies["cast"]+movies["crew"]
```

```python
movies.head()
```

```python
new_df=movies[["movie_id","title","tags"]]
```

```python
new_df.head()
```

```python
import warnings
warnings.filterwarnings("ignore")
new_df["tags"]=new_df["tags"].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df["tags"]=new_df["tags"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vector=cv.fit_transform(new_df["tags"]).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
similarity[0]
```

```python
def recommend(movie):
    movie_index=new_df[new_df["title"]==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(enumerate(distances),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```
